## Changing Columns DataTypes

```python
# changing timestamp
df['timestamp'] = pd.to_datetime(df['timestamp'])

# changing season
df['season'] = df['season'].astype(str)
```

```python
df.dtypes
```

```python
df
```

```python
# changing columns name
columns_name = {
    'timestamp': 'time',
    'cnt': 'count',
    't1': 'real_temp_C',
    't2': 'temp_feels_like',
    'hum': 'humidity_percent',
    'wind_speed': 'windspeed_kmh',
    'weather_code': 'weather',
    'is_holiday': 'is_holiday',
    'is_weekend': 'is_weekend',
    'season': 'season' 
}

df.rename(columns_name, axis='columns', inplace=True)
```

```python
# changing value humidity percent to desimal
df['humidity_percent'] = df['humidity_percent'] / 100
```

```python
df[df['season'] == '0.0']
```

```python
# changing season into string
change_season = {
    '0.0': 'spring',
    '1.0': 'summer',
    '2.0': 'fall',
    '3.0': 'winter'
}

df['season'] = df['season'].map(change_season)
```

```python
# changing weather values
df['weather'] = df['weather'].astype(str)
```

```python
df['weather'].value_counts()
```

```python
weather_dict = {
    '1.0': 'Clear',
    '2.0': 'Few clouds',
    '3.0': 'Broken clouds',
    '4.0': 'Cloudy',
    '7.0': 'Light rain',
    '10.0': 'Rain with tunderstorm',
    '26.0': 'Snowfall'
}

df['weather'] = df['weather'].map(weather_dict)
```
